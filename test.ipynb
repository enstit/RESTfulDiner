```python
from sqlalchemy import create_engine
from sqlalchemy.orm import Session

from models import Printer, Kiosk, Department, DeliveryStation, Item

engine = create_engine(
        "postgresql://postgres:postgres@srv-ud01:5432/postgres",
        connect_args={"options": "-csearch_path=sagrevolution"},
        echo=False,
    )

with Session(engine) as session:
    printers = session.query(Printer).all()
    print("### PRINTERS:")
    for printer in printers:
        print(printer)

    kiosks = session.query(Kiosk).all()
    print("\n### KIOSKS:")
    for kiosk in kiosks:
        print(kiosk)

    delivery_stations = session.query(DeliveryStation).all()
    print("\n### DELIVERY STATIONS:")
    for delivery_station in delivery_stations:
        print(delivery_station.to_dict())
        
    departments = session.query(Department).all()
    print("\n### DEPARTMENTS:")
    for department in departments:
        print(department)

    items = session.query(Item).all()
    print("\n### ITEMS:")
    for item in items:
        print(item)
```
